```python
def person_in_bounds(person):
    head = mean_head(person)
    if head[0]>= 0.4 and head[0]<= 0.6 and head[1]>=0.1 and head[1]<=0.4:
        return True
    else:
        return False
```

```python
def mean_head(person):
    head = []
    index = [0,1,14,15,16,17]
    x = []
    y = []
    for i in index:
        if i in person.body_parts:
            x.append(person.body_parts[i].x)
            y.append(person.body_parts[i].y)
    if len(x) ==0 :
        return None
    head.append(statistics.mean(x))
    head.append(statistics.mean(y))
    
    return head
```

```python
def mean_feet(person):
    feet = []
    index = [[10,13], [9,12], [8,11]]
    x = []
    y = []
    for ind in index:
        if len(x) == 0:
            for i in ind:
                if i in person.body_parts:
                    x.append(person.body_parts[i].x)
                    y.append(person.body_parts[i].y)
    if len(x) ==0 :
        return None
    feet.append(statistics.mean(x))
    feet.append(statistics.mean(y))
    
    return feet
```

```python
def distance(head,feet):
    return feet[1]-head[1]
```

```python
def find_batsman(first):
    batsman = ''
    max_distance = 0
    for i,person in enumerate(first):
        if person_in_bounds(person):
            head = mean_head(person)
            feet = mean_feet(person)
            if head is None or feet is None:
                return None
            d = distance(head,feet)
            if max_distance < d:
                max_distance = d
                batsman = person
    return batsman
```

```python
def mean_square_distance(person1,person2):
    distance = 0
    for parts in person1.body_parts:
        if parts in person2.body_parts:
            distance += ((person1.body_parts[parts].x - person2.body_parts[parts].x)**2 + (person1.body_parts[parts].y - person2.body_parts[parts].y)**2)
    min_parts = min(person1.part_count(), person2.part_count())
    if min_parts == 0:
        distance = 100
    else:
        distance = distance/min_parts
    return distance
        
```

```python
def find_next_batsmen(frame, person):
    distance = 100
    next_batsmen = person
    for f in frame:
        d = mean_square_distance(f,person)
        if distance > d:
            distance = d
            next_batsmen = f
    
    return next_batsmen
            
```

```python
def main(paths):
    batsmen_track = []
    try:
        first = pickle.load(open(paths[0], "rb" ))
        paths = paths[1:]
        person = find_batsman(first)
    except:
        save_path = output_path+str(paths[0]).split('/')[-2]+'_error.pkl'
        with open(save_path, 'wb') as f:
            pickle.dump(batsmen_track, f)
        return
        
        
    if person == '' or person is None:
        save_path = output_path+str(paths[0]).split('/')[-2]+'_error.pkl'
        with open(save_path, 'wb') as f:
            pickle.dump(batsmen_track, f)
        return
    batsmen_track.append(person)
    for p in paths:
        try:
            frame = pickle.load(open( p, "rb" ))
            person = find_next_batsmen(frame,person)
            batsmen_track.append(person)
        except:
            save_path = output_path+str(paths[0]).split('/')[-2]+'_error.pkl'
            with open(save_path, 'wb') as f:
                pickle.dump(batsmen_track, f)
            return
    
    save_path = output_path+str(paths[0]).split('/')[-2]+'.pkl'
    with open(save_path, 'wb') as f:
        pickle.dump(batsmen_track, f)
```
