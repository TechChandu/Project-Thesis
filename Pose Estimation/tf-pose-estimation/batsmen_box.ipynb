```python
import os
import pickle 
from pathlib import Path
import statistics 
import math

dirpath = '/home/chandu/Downloads/yolov5-master/misshit'
batsmen_path = '/home/chandu/Desktop/Pose Estimation/tf-pose-estimation/output/batsmen/misshit/'
parts = [8,11,4,7,9,12,3,6, 1,2,5,0,16,17,14,15,10,13]
output_path = '/home/chandu/Desktop/Pose Estimation/tf-pose-estimation/output/batsmen_box/misshit/'
paths = sorted(Path(dirpath).iterdir(), key=os.path.getmtime)
for path in paths:
    inner_dir = str(path)+'/labels'
    labels = sorted(Path(inner_dir).iterdir(), key=os.path.getmtime)
    b_path = batsmen_path+('_').join(str(labels[0]).split('/')[-1].split('.')[0].split('_')[:-1])+'.pkl'
    print(str(labels),b_path)
#     main(labels,b_path)
```

```python
def box_distance(batsmen,x):
    for p in parts:
        if batsmen.body_parts.get(p) is not None:
            return (batsmen.body_parts.get(p).x - x[0])**2 + (batsmen.body_parts.get(p).y - x[1])**2
```

```python
def main(labels,path):
    try:
        batsmen_dot = pickle.load(open(Path(path), "rb" ))
    except FileNotFoundError:
        save_path = output_path+str(path).split('/')[-1].split('.')[0] +'_error.pkl'
        with open(save_path, 'wb') as f:
            pickle.dump([], f)
        return
    
    batsmen_box = []
    if(len(batsmen_dot)<len(labels)-1):
        save_path = output_path+str(path).split('/')[-1].split('.')[0] +'_error.pkl'
        with open(save_path, 'wb') as f:
            pickle.dump([], f)
        return
    for index,l in enumerate(labels[:-1]):
        batsmen = batsmen_dot[index]
        with open(l,'r') as fp:
            Lines = fp.readlines()
            dis = 100
            box = ''
            for line in Lines:
                x = [float(i) for i in line.split()]
                if(x[0] == 0): # person
                    x = x[1:]
                    box_dis = box_distance(batsmen,x)
                    if(dis > box_dis):
                        dis = box_dis
                        box = x
        batsmen_box.append(box)
    
    save_path = output_path+str(path).split('/')[-1]
    with open(save_path, 'wb') as f:
        pickle.dump(batsmen_box, f)
        
```
